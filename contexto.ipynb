# Software Requirements Specification: CiudadAI

## 1. Visión General del Proyecto
CiudadAI es una plataforma inteligente para la gestión de incidencias municipales. [cite_start]Su objetivo es automatizar la clasificación y priorización de avisos ciudadanos mediante IA para mejorar la eficiencia operativa de los ayuntamientos[cite: 6, 8].

## 2. Definición del Producto
[cite_start]La aplicación se divide en dos secciones principales: el Portal Ciudadano (público) y el Panel de Gestión para Empleados (autenticado)[cite: 36, 38].

### 2.1 Portal Ciudadano (Frontend Público)
Interfaz simplificada para la interacción del ciudadano con el ayuntamiento.

* **Formulario de Creación de Incidencias:** Permite reportar problemas capturando los campos definidos en el dataset:
    * **Identificación Personal:** Nombre, Apellidos, NIF, Teléfono, Email, Dirección Personal.
    * **Detalles de la Incidencia:**
        * **Categoría (Desplegable):** Limpieza, Parques y Jardines, Movilidad, Mobiliario Urbano, Alumbrado Público y Otros.
        * **Descripción (Texto Libre):** Campo crítico que el modelo de IA procesará para determinar la urgencia.
        * **Ubicación:** Coordenadas (latitud/longitud) y descripción física del lugar.
        * **Fecha/Hora:** Captura automática del momento del reporte.
* **Consulta de Estado:**
    * Buscador por `ID_Incidencia`.
    * Visualización del estado actual: **Pendiente** o **Cerrado**.
* **Acceso de Empleados:** Botón visible para redirigir a la sección de autenticación.

### 2.2 Panel del Empleado (Frontend Autenticado)
Sección privada para la gestión técnica de los tickets.

* **Autenticación:** Acceso mediante Nombre de Usuario/Email y Contraseña.
* **Gestión de Tickets Asignados:**
    * **Visualización:** Lista de tickets asignados al empleado.
    * **Ordenación:** Prioridad obligatoria por nivel de **Urgencia** (Escala 1 a 5, siendo 5 la máxima prioridad) calculada por el modelo.
    * **Filtros por Estado:**
        * **Nuevo:** Ticket recién creado, no abierto.
        * **Pendiente:** Ticket en proceso de revisión.
        * **Cerrado:** Incidencia resuelta.
* **Resolución:**
    * Campo de texto libre para la respuesta del técnico.
    * Botón de envío que actualiza el estado a "Cerrado".

## 3. Especificaciones Técnicas (Stack Tecnológico)
* [cite_start]**Lenguaje:** Python[cite: 26].
* [cite_start]**IA/Machine Learning:** Scikit-Learn para el motor de clasificación de urgencia basado en la descripción[cite: 27].
* [cite_start]**Backend:** FastAPI para comunicación asíncrona[cite: 29].
* [cite_start]**Base de Datos:** PostgreSQL para almacenamiento persistente[cite: 28].
* [cite_start]**Infraestructura:** Docker (Contenedores) y GitHub (Control de versiones)[cite: 29, 30].

## 4. Requisitos de Datos y Seguridad
* [cite_start]**Dataset:** Uso de datos sintéticos generados con Tonic.ai para entrenamiento, incluyendo tipologías, metadatos temporales y niveles de prioridad[cite: 18, 20, 22].
* [cite_start]**Privacidad:** El sistema debe cumplir con el RGPD, implementando (opcionalmente) la anonimización de datos durante el proceso[cite: 6, 39].

## 5. Metodología de Desarrollo
* [cite_start]**Marco de Trabajo:** Scrum[cite: 32].
* [cite_start]**Ciclos:** Sprints de dos semanas con entregas de valor incrementales[cite: 33].
